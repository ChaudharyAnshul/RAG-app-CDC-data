```python
import pandas as pd
from ydata_profiling import ProfileReport
from sodapy import Socrata
```

```python
client = Socrata("data.cdc.gov", None)
```

```python
data = pd.read_csv("../data/cdc_data.csv", sep="\t")
```

```python
data
```

```python
# Ensure updateDate is in datetime format
data['updateDate'] = pd.to_datetime(data['updateDate'])
```

```python
date_threshold = pd.to_datetime('2021-01-01')
```

```python
# Filter the DataFrame
filtered_data = data[data['updateDate'] > date_threshold]
```

```python
filtered_data
```

```python
error = []
```

```python
def process_row(row):
  try:
    data_set_code = row['dataSetCode']
    data_set_type = row['dataSetType']
    title = row['title']

    if data_set_type not in ["Dataset", "Filtered View"]:
      return None

    df = pd.DataFrame.from_records(client.get(data_set_code))

    profile = ProfileReport(df, title=title, explorative=True)
    name = "reports/" + data_set_code
    profile.to_file(f"../{name}.html")

    return name
  except:
    print(f"ERROR: There is a error in processing row {title}")
    error.append(title)
    return None
```

```python
result_location = []
for _, row in filtered_data.iterrows():
  temp = process_row(row)
  result_location.append(temp)
```
